### Pivot Table

```python
import pandas as pd
```

```python
df=pd.read_csv('sales1.csv')
df
```

```python
df.pivot_table(index='type')
```

```python
df.pivot_table(index='type',columns='is_holiday')
```

```python
df.pivot_table(index='type',columns='is_holiday',aggfunc='count')
```

```python
df.pivot_table(index='type',columns='is_holiday',aggfunc='max')
```

```python
df.pivot_table(index='type',columns='is_holiday',aggfunc='sum')
```

```python
df.pivot_table(index='type',columns='is_holiday',fill_value='none')
```

```python
df.pivot_table(index='type',columns='is_holiday',fill_value='none',margins=True)
```

```python
df.pivot_table(index='type',columns='is_holiday',aggfunc='sum',fill_value='none',margins=True)
```
