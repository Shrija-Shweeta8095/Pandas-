### concat() function

```python
import pandas as pd
```

```python
s1=pd.Series([1,2,3,4])
s1
```

```python
s2=pd.Series([5,6,7,8])
s2
```

```python
pd.concat([s1,s2])
```

```python
s3=pd.Series([1,2,3])
s3
```

```python
pd.concat([s2,s3])
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                  'Name':['A','B','C','D'],
                  'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
                  'Name':['E','F','G','H'],
                  'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df2,df1],axis=1)
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                  'Name':['A','B','C','D'],
                  'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6],
                  'Name':['E','F'],
                  'Class':[9,10]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='outer')
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
                  'Name':['E','F','G','H'],
                  'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2],keys=['df1','df2'])
```

```python
pd.concat([df1,df2],keys=['First df','Second df'])
```

```python
pd.concat([df1,df2],axis=1,keys=['First df','Second df'])
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                  'Name':['A','B','C','D'],
                  'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'Marks':[40,34,56,91]})
df2
```

```python
pd.concat([df1,df2],sort=False)
```
