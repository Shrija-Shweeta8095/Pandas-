### melt() function

```python
import pandas as pd
```

```python
df=pd.read_csv('sales1.csv')
df
```

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars=['date'])
```

```python
pd.melt(df,id_vars=['department'])
```

```python
pd.melt(df,id_vars=['department'],value_vars=['unemployment'])
```

```python
pd.melt(df,id_vars=['department'],value_vars=['unemployment'],var_name='Category')
```
