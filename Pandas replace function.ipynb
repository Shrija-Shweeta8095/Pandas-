### Functions

#### replace()

```python
import pandas as pd
```

```python
df=pd.read_csv('advertisement.csv')
```

```python
df
```

```python
df.replace(to_replace=44.5,value=39)
```

```python
df.replace(to_replace=[22.1,10.4,9.3,22.0],value='Hi')
```

```python
df.replace({'TV':44.5},'None')
```

```python
df.replace({'TV':44.5,'Data':'No'},'None')
```

```python
df.replace('[A-Za-z]',0)
```

```python
df.replace('[A-Za-z]',0,regex=True)
```

```python
df.replace({'Data':'[A-Za-z]'},0,regex=True)
```

```python
df.replace(to_replace=44.5,method='ffill')
```

```python
df
```

```python
df.replace(to_replace=89,value=78,inplace=True)
```
